```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from siamese_net import ActionAutoencoder, CardAutoencoder, PseudoSiameseNet

%load_ext autoreload
%autoreload 2
```

```python

action_data = torch.rand(32, 24, 4, 9)  # 32 samples
card_data   = torch.rand(32, 6, 4, 13)   # 32 samples

action_dataset = TensorDataset(action_data)
card_dataset = TensorDataset(card_data)
action_loader = DataLoader(action_dataset, batch_size=8)
card_loader = DataLoader(card_dataset, batch_size=8)



# # Instantiate and train each autoencoder
action_autoencoder = ActionAutoencoder()
card_autoencoder = CardAutoencoder()

action_autoencoder.train_encoder(action_loader, num_epochs=200)
card_autoencoder.train_encoder(card_loader, num_epochs=200)

action_encoder = action_autoencoder.encoder
card_encoder = card_autoencoder.encoder 

siamese_net = PseudoSiameseNet(action_encoder, card_encoder)
```
